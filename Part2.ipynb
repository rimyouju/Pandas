#####
- 판다스 -> 다양한 형태의 외부 파일을 읽어와서 데이터프레임으로 변환하는 함수 제공
- 판다스 객체인 데이터프레임으로 변환되고 나면 판다스의 모든 함수와 기능을 자유롭게 사용 가능
- 데이터프레임을 다양한 유형의 파일으로 저장하는 것도 가능

#### CVS 파일

#####
- CVS(comma-seperated values)
- 데이터 값을 쉼표(,)로 구분하고 있음
- 쉼표(,)로 열을 구분 / 줄바꿈으로 행을 구분
- 판다스 read_csv() 함수에 확장자(.csv)를 포함하여 파일 경로(파일명)을 입력하면 CSV 파일을 읽어와서 데이터프레임으로 변환함

```python
# CSV파일 -> 데이터프레임
pandas.read_csv("파일 경로(이름)")
```

#####
- read_csv() 함수의 header 옵션은 데이터프레임의 열 이름으로 사용할 행을 지정
1. header = 0 (기본 값: 0행을 지정): df = read_csv(file)
2. header = 1 (1행을 열 지정): df = read_csv(file, header = 1)
3. header = None (행을 열 지정하지 않음): df = read_csv(file, header = None)

#####
- index_col 옵션은 데이터프레임의 행 인덱스가 되는 열을 지정
1. index_col = False (인덱스 지정하지 않음): df = read_csv(file, index_col = False)
2. index_col = 'c0' ('c0' 열을 인덱스 지정): df = read_csv(file, index_col = 'c0')

```python
# 2.1_csv_to_df.py

# 라이브러리 불러오기
import pandas as pd

# 파이썬 경로(파이썬 파일과 같은 폴더)를 찾고, 변수 file_path에 저장
file_path = '/Users/youju/Downloads/5674-833_4th/part2/read_csv_sample.csv'

# read_csv() 함수로 데이터프레임 변환. 변수 df1에 저장
df1 = pd.read_csv(file_path)
display(df1)
print('\n')

# read_csv() 함수로 데이터프레임 변환. 변수 df2에 저장. header = None 옵션
# 열 이름(행) 지정x -> 첫 행의 데이터가 열의 이름
df2 = pd.read_csv(file_path, header = None)
display(df2)
print('\n')

# read_csv() 함수로 데이터프레임 변환. 변수 df3에 저장. index_col = None 옵션
df3 = pd.read_csv(file_path, index_col = None) # 행 이름(열) 지정x -> 정수 0, 1, 2로 지정
display(df3)
print('\n')

# read_csv() 함수로 데이터프레임 변환. 변수 df4에 저장. index_col = 'c0' 옵션
df4 = pd.read_csv(file_path, index_col = 'c0')
display(df4)
```

#####
- csv 파일에 따라서 쉼표(,) 대신 탭(\t)이나 공백으로 텍스트를 구분하기도 함
- 구분자(sep 또는 delimiter) 옵션을 알맞게 입력해야 함

#####
- path -> 파일의 위치(파일명 포함), URL
- sep(또는 delimiter) -> 텍스트 데이터를 필드별로 구분하는 문자
- header -> 열 이름으로 사용될 행의 번호(기본 값은 0). header가 없고 첫 행부터 데이터가 있는 경우 None으로 지정 가능
- index_col -> 행 인덱스로 사용할 열의 변호 또는 열 이름
- names -> 열 이름으로 사용할 문자열의 리스트
- skiprows -> 처음 몇 줄을 skip할 것인지 설정(숫자 입력). skip하려는 행의 번호를 담은 리스트로 설정 가능 (예: [1, 3, 5])
- parse_dates -> 날짜 텍스트를 datetime64로 변환할 것인지 설정(기본값은 False)
- skip_footer -> 마지막 몇 줄을 skip할 것인지 설정(숫자 입력)
- encoding -> 덱스트 인코딩 종류를 지정

- 표 데이터들은 각각 별도의 데이터프레임으로 변환되기 때문에 여러 개의 데이터프레임(표)을 원소로 갖는 리스트가 반환됨

```python
# HTML 표 속성 읽기
pandas.read_html("웹 주소(URL)" 또는 "HTML 파일 경로(이름)")
```

```python
# 2.4_read_html.py

import pandas as pd

# HTML 파일 경로 or 웹 페이지 주소를 url 변수에 저장
url = '/Users/youju/Downloads/5674-833_4th/part2/sample.html'

# HTML 웹페이지의 표(table)을 가져와서 데이터프레임으로 변환
tables = pd.read_html(url)

# 표(table)의 개수 확인
print(len(tables))
print('\n')

# tables 리스트의 원소를 iteration하면서 각각 화면 출력
for i in range(len(tables)): # %s -> 문자열 대체
    print("tables[%s]" % i)
    display(tables[i])
    print(print('\n'))

# 파이썬 패키지 정보가 들어 있는 두 번째 데이터프레임을 선택하여 df 변수에 저장
df = tables[1]

# 'name' 열을 인덱스로 지정
df.set_index(['name'], inplace = True) # inplace = True -> df가 바로 변경됨
display(df)
```

#### 웹 스크래핑

#####
- 스크래핑 한 내용을 파이썬 리스트, 딕셔너리 등으로 정리한 뒤 DataFrame() 함수에 리스트나 딕셔너리 형태로 전달하여 데이터프레임으로 변환
- 딕셔너리 키는 열 이름 / 리스트는 열 데이터

```python
# 2.5_us_eff_list.py

# 라이브러리 불러오기
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd

# 위키피디아 미국 ETF 웹 페이지에서 필요한 정보를 스크래핑하여 딕셔너리 형태로 변수 etfs에 저장
url = "https://en.wikipedia.org/wiki/List_of_American_exchange-traded_funds"
resp = requests.get(url)
soup = BeautifulSoup(resp.text, 'lxml')
rows = soup.select('div > ul > li')

etfs = {}
for row in rows:

    try:
        etf_name = re.findall('^(.*) \(NYSE', row.text)
        etf_market = re.findall('\((.*)\|', row.text)
        etf_ticker = re.findall('NYSE Arca\|(.*)\)', row.text)

        if (len(etf_ticker) > 0) & (len(etf_market) > 0) & (len(etf_name) > 0):
            etfs[etf_ticker[0]] = [etf_market[0], etf_name[0]]

    except AttributeError as err:
        pass

#etfs 딕셔너리 출력
print(etfs)
print('\n')

# etfs 딕셔너리를 데이터프레임으로 변환
df = pd.DataFrame(etfs)
print(df)
```

```python
# 2.10_excewriter.py

import pandas as pd

# 판다스 DataFrame() 함수로 데이터프레임 변환. 변수 df1, df2에 저장
data1 = {'name' : ['Jarry', 'Riah', 'Paul'],
        'algol' : ["A", "A+", "B"],
        'basic' : ["C", "B", "B+"],
        'c++' : ["B+", "C", "C+"],
        }
data2 = {'c0' : [1, 2, 3],
         'c1' : [4, 5, 6],
         'c2' : [7, 8, 9],
         'c3' : [10, 11, 12],
         'c4' : [13, 14, 15]
         }

df1 = pd.DataFrame(data1)
df1.set_index('name', inplace = True) # name 열을 인덱스로 지정
display(df1)
print('\n')

df2 = pd.DataFrame(data2)
df2.set_index('c0', inplace = True) # c0 열을 인덱스로 지정
display(df2)

# df1을 'sheet1'으로, df2를 'sheet2'로 저장(Excel 파일명은 "df_excelwriter.xlsx")
writer = pd.ExcelWriter("./df_excelwriter.xlsx")
df1.to_excel(writer, sheet_name = "sheet1")
df2.to_excel(writer, sheet_name = "sheet2")
writer.save()
```
