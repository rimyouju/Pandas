### 데이터프레임 합치기

- 데이터가 여러 군데 나누어져 있을 때 하나로 합치거나 데이터를 연결해야 하는 경우
- concat(), merge(), join()

#### 데이터프레임 연결

- 서로 다른 데이터프레임들의 구성 형태와 속성이 균일하다면, 행 또는 열 중에 어느 한 방향으로 이어 붙여도 데이터의 일관성을 유지할 수 있음
- 기존 데이터프레임의 형태를 유지하면서 이어 붙이는 개념으로 판다스 count() 함수 활용

```python
# 데이터프레임 연결
pandas.count(데이터프레임의 리스트)
```

- count() 함수에 데이터프레임을 원소로 갖는 리스트를 전달하면 여러 개의 데이터프레임을 서로 연결
- 축방향을 지정하지 않으면 기본 옵션(axis = 0)이 적용되어 위 아래 행 방향으로 연결
- 각 데이터프레임의 행 인덱스는 본래 형태를 유지
- 열 이름에 대해서는 join = 'outer' 옵션이 기본 적용         
-> 두 열 이름의 합집합으로 연결 데이터프레임의 열 이름 배열을 구성
- join = 'inner' 옵션의 경우 데이터프레임에 공통으로 속하는 교집합이 기준

```python
# 6.11_df_concat.py

# 라이브러리 불러오기
import pandas as pd

# 데이터프레임 만들기
df1 = pd.DataFrame({'a' : ['a0', 'a1', 'a2', 'a3'],
                    'b' : ['b1', 'b2', 'b3', 'b4'],
                    'c' : ['c1', 'c2', 'c3', 'c4']},
                    index = [0, 1, 2, 3])
df2 = pd.DataFrame({'a' : ['a2', 'a3', 'a4', 'a5'],
                    'b' : ['b2', 'b3', 'b4', 'b5'],
                    'c' : ['c2', 'c3', 'c4', 'c5'],
                    'd' : ['d2', 'd3', 'd4', 'd5']},
                    index = [2, 3, 4, 5])

print(df1, '\n')
print(df2, '\n')

# 2개의 데이터프레임을 위 아래 행 방향으로 이어 붙이듯 연결하기
result1 = pd.concat([df1, df2])
print(result1, '\n')
```

```python
# ignore_index = True 옵션 설정 
# -> 기존 행 인덱스를 무시하고 새로운 행 인덱스 설정
result2 = pd.concat([df1, df2], ignore_index = True)
print(result2)
```

- axis = 1 옵션을 사용하면 데이터프레임을 좌우 열 방향으로 연결
- 기존 열 이름 배열이 그대로 유지
- 연결되는 데이터프레임의 행 인덱스는, join = 'outer' 옵션이 기본값으로 적용되어 각 데이터프레임의 행 인덱스들의 합집합으로 고성

```python
# 2개의 데이터프레임을 좌우 열 방향으로 이어 붙이듯 연결
result3 = pd.concat([df1, df2], axis = 1)
print(result3, '\n')
```

- join = 'inner' 옵션을 설정하면 연결할 데이터프레임들의 행 인덱스의 교집합을 기준으로 사용

```python
# join = 'inner' 옵션 적용하기(교집합)
result3_in = pd.concat([df1, df2], axis = 1, join = 'inner')
print(result3_in, '\n')
```

- 데이터프레임과 시리즈를 좌우 열 방향으로 연결도 가능 => 데이터프레임에 열을 추가하는 것과 같음
- 시리즈의 이름의 데이터프레임의 열 이름으로 반환
- 데이터프레임의 행 인덱스와 시리즈의 인덱스가 같아야 함
- 공통 인덱스가 없을 경우 NaN 값이 입력됨

```python
# 시리즈 만들기
sr1 = pd.Series(['e0', 'e1', 'e2', 'e3'], name = 'e')
sr2 = pd.Series(['f0', 'f1', 'f2'], name = 'f', index = [3, 4, 5])
sr3 = pd.Series(['g0', 'g1', 'g2', 'g3'], name = 'g')

# df1과 sr1을 좌우 열 방향으로 연결하기
result4 = pd.concat([df1, sr1], axis = 1)
print(result4, '\n')

# df2와 sr2를 좌우 열 방향으로 연결하기
result5 = pd.concat([df2, sr2], axis = 1, sort = True)
print(result5)
```

- 시리즈들로 만든 리스트를 concat() 함수에 전달하면 시리즈가 서로 연결됨
- axis = 1 옵션을 적용하면 좌우 열 방향으로 연결하여 데이터프레임이 됨
- axis = 0 옵션을 적용하면 위 아래 방향으로 길게 연결되어 하나의 시리즈가 됨

```python
# sr1과 sr3을 좌우 열 방향으로 연결하기
result6 = pd.concat([sr1, sr3], axis = 1)
print(result6, '\n')

result7 = pd.concat([sr1, sr3], axis = 0)
print(result7, '\n')
```

- on = 'id' 옵션은 두 데이터프레임의 공통 열 중에서 'id' 열을 키로 병합
- how = 'outer' 옵션은 기준이 되는 'id' 열의 데이터가 데이터프레임 중 어느 한쪽에만 속하더라도 포함

- how = 'left' 옵션을 설정하면 왼쪽 데이터프레임의 키 열에 속하는 데이터 값을 기준으로 병합

- 오른쪽 데이터프레임의 키 열을 기준으로 추출

```python
# 6.14_df_groupby.py

# 라이브러리 불러오기
import pandas as pd
import seaborn as sns

# titanic 데이터셋에서 age, sex 등 5개 열을 선택하여 데이터프레임 만들기
titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age', 'sex', 'class', 'fare', 'survived']]

print('승객 수:', len(df))
print(df.head())
print('\n')

# class 열을 기준으로 분할
grouped = df.groupby(['class'])
print(grouped)
```

```python
# 연산 메서드 적용
average = grouped.mean()
print(average)
```

```python
# 개별 그룹 선택하기
group3 = grouped.get_group('Third')
print(group3.head())
```

#### 그룹 연산 메서드(적용-결합 단계)

##### 데이터 집계

- 그룹 객체에 다양한 연산 적용 가능

```python
# 표준편차 데이터 집계(내장함수)
group객체.std()
```

```python
# 6.15_groupby_agg.py

# 라이브러리 불러오기
import pandas as pd
import seaborn as sns

# titanic 데이터셋에서 age, sex 등 5개 열을 선택하여 데이터프레임 만들기
titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age', 'sex', 'class', 'fare', 'survived']]

# class 열을 기준으로 분할
grouped = df.groupby(['class'])

# 각 그룹에 대한 모든 열의 표준편차를 집계하여 데이터프레임으로 반환
std_all = grouped.std()
print(std_all)
print('\n')
print(type(std_all))
print('\n')

# 각 그룹에 대한 fare 열의 표준편차를 집계하여 시리즈로 반환
std_fare = grouped.fare.std()
print(std_fare)
print('\n')
print(type(std_fare))
```

- 집계 연산을 처리하는 사용자 정의 함수를 그룹 객체에 적용하려면 agg() 메서드를 사용

```python
# agg() 메서드 데이터 집계
group객체.agg(매핑 함수)
```

```python
# 그룹 객체에 agg() 메서드 적용 - 사용자 정의 함수를 인자로 전달
def min_max(x): # 최댓값 - 최솟값
    return x.max() - x.min()

# 각 그룹의 최댓값과 최솟값의 차이를 계산하여 그룹별로 집계
agg_minmax = grouped.agg(min_max)
print(agg_minmax.head())
```

- 동시에 여러 개의 함수를 사용하여 각 그룹별 데이터에 대한 집계 연산 처리 가능
- 각각의 열에 여러 개의 함수를 일괄 적용할 때는 리스트 형태로 인수를 전달
- 열마다 다른 종류의 함수를 적용하려면 {열 : 함수} 형태의 딕셔너리를 전달

```python
# 모든 열에 여러 함수를 매핑
group객체.agg([함수1, 함수2, 함수3,...])

# 각 열마다 다른 함수를 매핑
group객체.agg({'열1' : 함수1, '열2' : 함수2})
```

```python
# 여러 함수를 각 열에 동일하게 적용하여 집계
agg_all = grouped.agg(['min', 'max'])
print(agg_all.head())
print('\n')

# 각 열마다 다른 함수를 적용하여 집계
agg_sep = grouped.agg({'fare' : ['min', 'max'], 'age' : 'mean'})
print(agg_sep.head())
```

- 2개의 함수를 리스트 형태로 입력하면 각 열에 대하여 2개 함수의 연산 결과를 각각 집계하여 다른 열로 구분하여 표시
- 함수명을 열 이름에 추가하여 2중 열 구조를 만든다

##### 그룹 연산 데이터 변환

- agg(): 각 그룹별 데이터에 연산을 위한 함수를 구분 적용; 그룹별로 연산 결과를 집계하여 반환
- transform(): 그룹별로 구분하여 각 원소에 함수를 적용; 각 원소의 본래 행 인덱스와 열 이름을 기준으로 연산 결과 반환
- 그룹 연산의 결과를 원본 데이터프레임과 같은 형태로 변형하여 정리

```python
# 데이터 변환 연산
group객체.transform(매핑 함수)
```

```python
# 6.16_df_groupby_transform.py

# 생략 (6.15와 동일)

# 그룹별 age 열의 평균 집계 연산
age_mean = grouped.age.mean()
print(age_mean)
print('\n')

# 그룹별 age 열의 표준편파 집계 연산
age_std = grouped.age.std()
print(age_std)
print('\n')

# 그룹 객체의 age 열을 iteration으로 z-score를 계산하여 출력
for key, group in grouped.age:
    group_zscore = (group - age_mean.loc[key]) / age_std.loc[key]
    print('* origin:', key)
    print(group_zscore.head(3))
    print('\n')
```

- transform() 메서드를 사용하여 z-score로 직접 변환
- z-score를 계산하는 사용자 함수를 정의하고, transform() 메서드의 인자로 전달
- 각 그룹별 평균과 표준편차를 이용하여 각 원소의 z-score를 계산
- 반환되는 객체는 그룹별로 나누지 않고 원래 행 인덱스 순서로 정렬됨

```python
# z-score를 계산하는 사용자 함수 정의
def z_score(x):
    return (x - x.mean()) / x.std()

# transforn() 메서드를 이용하여 age 열의 데이터를 z-score로 변환
age_zscore = grouped.age.transform(z_score)
print(age_zscore.loc[[1, 9, 0]]) # 1, 2, 3 그룹의 첫 데이터 확인(변환 결과)
print('\n')
print(len(age_zscore)) # transform 메서드 반환 값의 길이
print('\n')
print(age_zscore.loc[0:9]) # transform 메서드 반환 값 출력(첫 10개)
print('\n')
print(type(age_zscore)) # transform 메서드 반환 객체의 자료형
```

### 피벗

- 판다스 pivot_table() 함수는 엑셀에서 사용하는 피벗테이블과 비슷한 기능을 처리
- 피벗테이블을 구성하는 4가지 요소(행 인덱스, 열 인덱스, 데이터 값, 데이터 집계 함수)에 적용할 데이터프레임의 열을 각각 지정하여 함수의 인자로 전달

```python
# 6.20_pivot_table.py

# 라이브러리 불러오기
import pandas as pd
import seaborn as sns

# IPython 디스플레이 설정 변경
pd.set_option('display.max_columns', 10) # 출력할 최대 열의 개수
pd.set_option('display.max_colwidth', 20) # 출력할 열의 너비

# titanic 데이터셋에서 age, sex 등 5개 열을 선택하여 데이터프레임 만들기
titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age', 'sex', 'class', 'fare', 'survived']]
print(df.head())
print('\n')

# 행, 열, 값, 집계에 사용할 열을 1개씩 지정 - 평균 집계
pdf1 = pd.pivot_table(df, # 피벗할 데이터프레임
                      index = 'class', # 행 위치에 들어갈 열
                      columns = 'sex', # 열 위치에 들어갈 열
                      values = 'age',  #데이터로 사용할 열
                      aggfunc = 'mean') # 데이터 집계 함수

print(pdf1.head())
```

```python
# 값에 적용하는 집계 함수 2개 이상 지정 가능 - 생존율, 생존자 수 집계
pdf2 = pd.pivot_table(df, # 피벗할 데이터프레임
                      index = 'class', # 행 위치에 들어갈 열
                      columns = 'sex', # 열 위치에 들어갈 열
                      values = 'survived', # 데이터로 사용할 열
                      aggfunc = ['mean', 'sum']) # 데이터 집계 함수

print(pdf2.head())
```

- pivot_table() 함수에 전달할 인자(행 인덱스, 열 위치, 데이터 값, 집계 함수)에 각각 2개 이상의 열 입력 가능

```python
# 행, 열, 값에 사용할 열을 2개 이상 지정 가능 - 평균 나이, 최대 요금 집계
pdf3 = pd.pivot_table(df, # 피벗할 데이터프레임
                      index = ['class', 'sex'], # 행 위치에 들어갈 열
                      columns = 'survived', # 열 위치에 들어갈 열
                      values = ['age', 'fare'], # 데이터로 사용할 열
                      aggfunc = ['mean', 'max']) # 데이터 집계 함수

# IPython Console 디스플레이 옵션 설정
pd.set_option('display.max_columns', 10) # 출력할 열의 최대 개수
print(pdf3.head())
print('\n')

# 행, 열 구조 살펴보기
print(pdf3.index)
print(pdf3.columns)
```

- index 속성과 columns 속성에는 3가지 속성이 있음
- levels 속성을 보면 행 인덱스는 2개의 열이 사용되어 2중 구조
- 열은 3중 구조로 이루어짐

- xs 인덱서는 기본값으로 행 인덱스에 접근하고, 축 값은 axis = 0으로 자동 설정

```python
# xs 인덱서 사용 - 행 선택(default: axis = 0)
print(pdf3.xs('First')) # 행 인덱스가 First인 행을 선택
```

```python
print(pdf3.xs(('First', 'female'))) # 행 인덱스가 ('First', 'female')인 행을 선택
```

```python
# 행 인덱스 레벨 직접 지정
print(pdf3.xs('male', level = 'sex')) # 행 인덱스의 sex 레벨이 male인 행을 선택
```

```python
print(pdf3.xs(('Second', 'male'), level = [0, 'sex'])) # Second, male인 행을 선택
```

```python
# xs 인덱서를 이용하여 열 인덱스에 접근하기 위해서는 축 값을 axis = 1로 설정
# xs 인덱서 사용 - 열 선택(axis = 1 설정)
print(pdf3.xs('mean', axis = 1)) # 열 인덱스가 mean인 데이터 선택
```

```python
# 그룹별 집계 결과만을 추출하여 반환
print(pdf3.xs(('mean', 'age'), axis = 1)) # 열 인덱스가 ('mean', 'age')인 데이터 선택
```

```python
# 열 인덱스 레벨 직접 지정
print(pdf3.xs(1, level = 'survived', axis = 1)) # survived 레벨이 1인 데이터 선택
```

```python
print(pdf3.xs(('max', 'fare', 0),
              level = [0, 1, 2], axis = 1)) # max, fare, survived = 0인 데이터 선택
```
