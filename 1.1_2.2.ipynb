## 판다스 자료구조

##### 
- 분석을 위해 다양한 sorce로부터 수집하는 데이터는 형태나 속성이 매우 다양함  
-> 여러 종류의 데이터를 컴퓨터가 이해할 수 있도록 동일한 형식을 갖는 구조로 통합해야 함

### 시리즈(Series)와 데이터프레임(DataFrame)

##### 서로 다른 종류의 데이터를 한 곳에 담는 그릇(컨테이너)
- 시리즈 -> 1차원 배열
- 데이터프레임 -> 2차원 배열

##### 판다스의 목적
- 서로 다른 여러 가지 유형의 데이터를 공통의 포멧으로 정리하는 것
특히 행과 열로 이루어진 2차원 구조의 데이터프레임은 데이터 분석 실무에서 자주 사용됨

### 시리즈

##### 
- 데이터가 순차적으로 나열된 1차원 배열의 형태를 가짐
- 인덱스(index)는 데이터 값(value)와 일대일 대응됨 
-> 딕셔너리와 비슷한 구조

#### 시리즈 만들기
##### 판다스 내장 함수인 Series()이용하고, 딕셔너리를 함수의 인자로 전달

```python
# 딕셔너리 -> 시리즈 변환: 
pandas.Series(딕셔너리)
```

##### 
- 딕셔너리의 키(key)는 시리즈의 인덱스에 대응, 
- 딕셔너리의 각 키에 매칭되는 값(value)이 시리즈의 데이터 값으로 변환

```python
# -*- coding: utf-8 -*-

# 1.1_dict_to_series.py

# pandas 불러오기
import pandas as pd

# ketLvalue 쌍으로 딕셔너리를 만들고, 변수 dict_data에 저장
dict_data = {'a' : 1, 'b' : 2, 'c': 3}

# 판다스 Series() 함수로 dictionary를 Series로 변환, 변수 sr에 저장
sr = pd.Series(dict_data)

# sr의 자료형 출력
print(type(sr))
print('\n')
# 변수 sr에 저장되어 있는 시리즈 객체 출력
print(sr)
```

##### 
- sr에 저장된 객체의 자료형 -> Series 클래스
- 인덱스 'a', 'b', 'c'는 왼쪽, 값 '1', '2', '3'은 오른쪽 표시
- 시리즈를 구성하는 데이터 값의 자료형(dtype)은 정수형(int64)

#### 인덱스 구조

##### 
- 인덱스는 자기와 짝을 이루는 데이터 값의 순서와 주소를 저장함       
-> 데이터의 탐색, 정렬, 선택, 겨합 등 데이터 조작을 쉽게 할 수 있음

- 인덱스는 정수형 위치 인덱스(integer position), 인덱스 이름(index name) 또는 인덱스 라벨(index label) 두 가지 종류가 있음

```python
# 시리즈 클래스의 index 속성을 이용하여 인덱스 배열을 따로 선택할 수 있음
# 인덱스 배열
Series객체.index

# 시리즈 클래스의 values 속성을 이용하여 데이터 값 배열을 따로 선택할 수 있음
# 데이터 값 배열
Series객체.values
```

단, 리스트를 시리즈로 변환할 때는 딕셔너리의 키처럼 인덱스로 변환될 값이 없음    
-> 인덱스를 별도로 정의하지 않으면, 디폴트로 정수형 위치 인덱스가 자동으로 지정됨

```python
# 1.2_series_index.py

import pandas as pd

# 리스트를 시리즈로 변환하여 변수 sr에 저장
list_data = ['2019-01-02', 3.14, 'ABC', 100, True]
sr = pd.Series(list_data)
print(sr)
```

#####
- 시리즈의 index 속성과 values 속성을 이용하면 인덱스 배열과 데이터 값의 배열을 불러올 수 있음

#### 데이터프레임

#####
- 2차원 배열(행과 열)
- 열 -> 각각 시리즈 객체
- 시리즈를 열벡터(vector)라고 하면, 데이터프레임은 여러 개의 열벡터들이 같은 행 인덱스를 기준으로 줄지어 결합된 2차원 벡터 또는 행렬(matrix)

##### 
- 데이터프레임은 행과 열을 나타내기 위해 두 가지 종류의 주소를 사용
- 인덱스(row index)와 열 이름(column name 또는 column label)으로 구분

##### 
- 데이터 프레임 열 -> 공통의 속성을 갖는 일련의 데이터(변수)
- 행 -> 하나의 주식 종목에 관한 관측값(öbservatiön)

#### 데이터프레임 만들기

##### 
- 데이터 프레임을 만들이 위해서는 같은 길이(원소의 개수가 동일한)의 1차원 배열이 여러 개가 필요         
-> 여러 개의 시리즈(열, cölumn)를 모아 놓은 집합

##### 
- 딕셔너리의 값(value)에 해당하는 각 리스트가 시리즈 배열로 변환되어 데이터 프레임의 열이 된다

- 딕셔너리의 키(key)는 각 시리즈의 이름으로 변환되어 최종적으로 데이터프레이므이 열 이름이 된다

```python
# 데이터프레임을 만들 때는 판다스 DataFrame() 함수를 사용
# 여러 개의 리스트를 원소로 갖는 딕셔너리를 함수의 인자로 전달하는 방식이 주로 사용
# 딕셔너리 -> 데이터프레임 변환
pandas.DataFrame(딕셔너리 객체)
```

```python
# 1.4_dict_to_dataframe.py

import pandas as pd

# 열 이름을 key로 하고, 리스트를 value로 갖는 딕셔너리 정의(2차원 배열)
dict_data = {'c0' : [1, 2, 3], 'c1' : [4, 5, 6], 'c2' : [7, 8, 9], 'c3' : [10, 11, 12], 'c4' : [13, 14, 15]}

# 판다스 DataFrame() 함수로 딕셔너리를 데이터프레임으로 변환. 변수 df에 저장
df = pd.DataFrame(dict_data)

# df의 자료형 출력
print(type(df))
print('\n')
# 변수 df에 저장되어 있는 데이터프레임 객체를 출력
print(df)
```

##### 
- 딕션너리의 키(k) -> 열 이름(c0~c4)
- 값(v) -> 데이터프레임의 열
- 행 인덱스에는 정수형 위치 인덱스가 자동 지정
- 반환되는 객체는 데이터프레임

```python
# df.index와 df.columns 속성에 새로운 배열을 할당하는 방식으로 행 인덱스와 열 이름을 변경할 수 있음
# 행 인덱스 변경
DataFrame객체.index = 새로운 행 인덱스 배열

# 열 이름 변경
DataFrame객체.columns = 새로운 열 이름 배열
```

```python
# 1.5_change_df_idx_col.py


import pandas as pd

# 행 인덱스/열 이름 지정하여 데이터프레임 만들기
df = pd.DataFrame([[15, '남', '덕영중'], [17, '여', '수리중']],
                  index = ['준서', '예은'],
                  columns = ['나이', '성별', '학교'])

# 행 인덱스, 열 이름 확인
print(df) # 데이터프레임
print('\n')
print(df.index) # 행 인덱스
print('\n')
print(df.columns) # 열 이름

# 행 인덱스, 열 이름 변경
df.index = ['학생1', '학생2']
df.columns = ['연령', '남녀', '소속']

print(df) # 데이터프레임
print('\n')
print(df.index) # 행 인덱스
print('\n')
print(df.columns) # 열 이름
```

#####
- df의 행 인덱스를 ['준서', '예은']에서 ['학생1', '학생2']로 변경
- 열 이름은 ['나이', '성별', '학교']에서 ['연령', '남녀', '소속']으로 변경

##### 
- 데이터프레임에 rename( ) 메소드를 적용하면 행 인덱스 또는 열 이름의 일부를 선택하여 변경 할 수 있음      
-> 원본 객체를 직접 수정하는 것이 아니라 새로운 데이터프레임 객체를 반환
- 원본 객체를 변경하려면 inplace=True 옵션을 사용

```python
# 행 인덱스 변경
DataFrame객체.rename(index = {기존행인덱스1 : 새로운행인덱스1,
                                기존행인덱스2 : 새로운행인덱스2,
                                ...},
                       inplace = True/False)

# 열 이름 변경
DataFrame객체.rename(columns = {기존열이름1 : 새로운열이름1,
                                 기존열이름2 : 새로운열이름2,
                                 ...},
                      inplace = True/False)
```

```python
# 1.6_change_df_idx_col2.py

import pandas as pd

# 행 인덱스/열 이름 지정하여 데이터프레임 만들기
df = pd.DataFrame([[15, '남', '덕영중'], [17, '여', '수리중']],
                  index = ['준서', '에은'],
                  columns = ['나이', '성별', '학교'])

# 데이터프레임 df 출력
print(df)
print('\n')

# 열 이름 중, '나이'를 '연령'으로, '성별' : '남녀', '학교' : '소속'으로 바꾸기
df.rename(columns = {'나이' : '연령', '성별' : '남녀', '학교' : '소속'}, inplace = True)

# df의 행 인덱스 중에서, '준서'를 '학생1'로, '예은'을 '학생2'로 바꾸기
df.rename(index = {'준서' : '학생1', '예은' : '학생2'}, inplace = True)

# df 출력(변경 후)
print(df)
```

#####
- df의 행 인덱스를 ['준서', '예은']에서 ['학생1', '학생2']로 변경
- 열 이름은 ['나이', '성별', '학교']에서 ['연령', '남녀', '소속']으로 변경

#### 행/열 삭제

#####
- 데이터프레임의 행 또는 열을 삭제하는 명령으로 drop() 메서드 사용
- 행을 삭제할 때는 축 (axis) 옵션으로 axis=0을 입력하거나, 별도로 입력하지 않는다
- 반면, 축 옵션으로 axis=1을 입력하면 열을 삭제
- 동시에 여러 개의 행 또는 열을 삭제하려면 리스트 형태로 입력

#####
- drop() 메소드는 기존 객체를 변경하지 않고 새로운 객체를 반환
- 원본 객체를 직접 변경하기 위해서는 inplace=True 옵션을 추가
- inplace 옵션을 False로 설정하면 새로운 객체가 생성되고, 원 객체는 변경 없이 그대로 유지된다 


```python
# 행 삭제
DataFrame객체.drop(행 인덱스 또는 배열, axis = 0)

# 열 삭제
DataFrame객체.drop(열 이름 또는 배열, axis = 1)
```

```python
# 1.9_select_row.py

import pandas as pd

# DataFrame() 함수로 데이터프레임 변환. 변수 df에 저장
exam_data = {'수학' : [90, 80, 70], '영어' : [98, 89, 95],
             '음악' : [85, 95, 100], '체육' : [100, 95, 90]}

df = pd.DataFrame(exam_data, index = ['서준', '우현', '인아'])
print(df)
print('\n')

# 헹 인덱스를 사용하여 행 1개 선택
lable1 = df.loc['서준']
position1 = df.iloc[0]
print(lable1)
print('\n')
print(position1)
```

```python
# 1.14_modify_dataframe_element.py
# 각기 다른 점수를 새로운 값으로 입력하여 원소 변경

import pandas as pd

# DataFrame() 함수로 데이터프레임 변환. 변수 df에 저장
exam_data = {'이름' : ['서준', '우현', '인아'],
             '수학' : [90, 80, 70], 
             '영어' : [98, 89, 95],
             '음악' : [85, 95, 100], 
             '체육' : [100, 95, 90]}
df = pd.DataFrame(exam_data)

# '이름' 열을 새로운 인덱스로 지정하고, df 객체에 변경사항 반영
df.set_index('이름', inplace = True)
print(df)
print('\n')

# 데이터프레임 df의 특정 원소를 변경하는 방법: '서준'의 '체육' 점수
df.iloc[0][3] = 80
print(df)
print('\n')

df.loc['서준']['체육'] = 90
print(df)
print('\n')

df.loc['서준', '체육'] = 100
print(df)
```

```python
# 여러 개의 원소를 선택하여 새로운 값을 할당하면, 모든 원소를 한꺼번에 같은 값으로 변경할 수 있음
# 데이터프레임 df의 원소 여러 개를 변경하는 방법: '서준'의 '음악', '체육' 점수
df.loc['서준', ['음악', '체육']] = 50
print(df)
print('\n')

df.loc['서준', ['음악', '체육']] = 100, 50
print(df)
```

#####
전치의 결과로 새로운 객체를 반환하므로, 기존 객체를 변경하기 위해서는 df = df.transpose() 또는 df = df.T와 같이 기존 객체에서 새로운 객체를 할당해주는 과정이 필요

```python
# 행, 열 바꾸기
DataFrame객체.transpose()
DataFrame객체.T
```

```python
# 1.15_transpose.py

import pandas as pd

# DataFrame() 함수로 데이터프레임 변환. 변수 df에 저장
exam_data = {'이름' : ['서준', '우현', '인아'],
             '수학' : [90, 80, 70], 
             '영어' : [98, 89, 95],
             '음악' : [85, 95, 100], 
             '체육' : [100, 95, 90]}
df = pd.DataFrame(exam_data)
print(df)
print('\n')

# 데이터프레임 df를 전치하기(메서드 활용)
df = df.transpose()
print(df)
print('\n')

# 데이터프레임 df를 다시 전치하기(클래스 속성 활용)
df = df.T
print(df)
```

#####
- transpose() 메소드를 적용하여 행과 열을 서로 바꾸는 전치 과정
- 열 이름에 해당하는 과목명이 행 인덱스 위치로 이동하고, 행 인덱스에 있는 숫자들이 열 이름으로 이동
- df.T와 같이 클래스 속성을 활용할 수도 있음
- 여기서 데이터프레임의 행과 열을 한 번 더 바꾸면, 최초의 원본 데이터프레임과 같은 형태로 돌아옴
